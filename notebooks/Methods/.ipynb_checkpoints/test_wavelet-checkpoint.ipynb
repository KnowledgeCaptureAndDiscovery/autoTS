# Wavelet analysis

The goal of this notebook is to compare the continuous wavelet transform as implemented in [Pywavelets](https://pywavelets.readthedocs.io/en/latest/ref/cwt.html?highlight=cwt) with the weighted wavelet Z transform.

```python
from pyleoclim.utils import wavelet as waveutils
import pyleoclim as pyleo
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

## Wavelet methods

### Default scale vector

Use the method in Torrance and Compo to obtain wave from frequency for a Morlet wavelet. See [here](https://github.com/chris-torrence/wavelets/blob/master/wave_python/waveletFunctions.py)


```python
def freq2scale(freq):
    fourier_factor = 4*np.pi/(6+np.sqrt(2+6**2))
    scales=1./(fourier_factor*freq)
    return scales
```

## New proposed Pyleoclim UI function for Class Series.

Not that the argument would remain the same. 

```python
def wavelet(ts,method='wwz',settings=None,freq_method='log', ntau=None,freq_kwargs=None,verbose=False):
    
    if not verbose:
        warnings.simplefilter('ignore')

    settings = {} if settings is None else settings.copy()
    wave_func = {
        'wwz': waveutils.wwz,
        'cwt': waveutils.cwt,
    }

    if method == 'cwt' and 'freq' in settings.keys():
        scales=freq2scale(freq)
        settings.update({'scales':scales})
        del settings['freq']

    freq_kwargs = {} if freq_kwargs is None else freq_kwargs.copy()
    freq = waveutils.make_freq_vector(ts.time, method=freq_method, **freq_kwargs)

    args = {}

    if ntau is None:
        ntau = np.min([np.size(ts.time), 50])

    tau = np.linspace(np.min(ts.time), np.max(ts.time), ntau)

    args['wwz'] = {'tau': tau, 'freq': freq}
    args['cwt'] = {'wavelet' : 'morl', 'scales':freq2scale(freq)}


    args[method].update(settings)
    wave_res = wave_func[method](ts.value, ts.time, **args[method])
    scal = pyleo.Scalogram(
        frequency=wave_res.freq,
        time=wave_res.time,
        amplitude=wave_res.amplitude,
        coi=wave_res.coi,
        label=ts.label,
        timeseries=ts,
        wave_method=method,
        freq_method=freq_method,
        freq_kwargs=freq_kwargs,
        wave_args=args[method],
    )

    return scal  
```

```python
def compare_method(ts):
    scal_wwz = wavelet(ts)
    scal_cwt=wavelet(ts,method='cwt',settings={'method':'fft'},freq_method='lomb_scargle')
    fig1=scal_wwz.plot()
    fig2=scal_cwt.plot()
    return fig1, fig2
```

```python
def colored_noise(alpha, t, f0=None, m=None, seed=None):
    ''' Generate a colored noise timeseries
    Args:
        alpha (float): exponent of the 1/f^alpha noise
        t (float): time vector of the generated noise
        f0 (float): fundamental frequency
        m (int): maximum number of the waves, which determines the
            highest frequency of the components in the synthetic noise
    Returns:
        y (array): the generated 1/f^alpha noise
    References:
        Eq. (15) in Kirchner, J. W. Aliasing in 1/f(alpha) noise spectra: origins, consequences, and remedies.
            Phys Rev E Stat Nonlin Soft Matter Phys 71, 066110 (2005).
    '''
    n = np.size(t)  # number of time points
    y = np.zeros(n)

    if f0 is None:
        f0 = 1/n  # fundamental frequency
    if m is None:
        m = n//2

    k = np.arange(m) + 1  # wave numbers

    if seed is not None:
        np.random.seed(seed)

    theta = np.random.rand(int(m))*2*np.pi  # random phase
    for j in range(n):
        coeff = (k*f0)**(-alpha/2)
        sin_func = np.sin(2*np.pi*k*f0*t[j] + theta)
        y[j] = np.sum(coeff*sin_func)

    return y

def colored_noise_2regimes(alpha1, alpha2, f_break, t, f0=None, m=None, seed=None):
    ''' Generate a colored noise timeseries with two regimes
    Args:
        alpha1, alpha2 (float): the exponent of the 1/f^alpha noise
        f_break (float): the frequency where the scaling breaks
        t (float): time vector of the generated noise
        f0 (float): fundamental frequency
        m (int): maximum number of the waves, which determines the
            highest frequency of the components in the synthetic noise
    Returns:
        y (array): the generated 1/f^alpha noise
    References:
        Eq. (15) in Kirchner, J. W. Aliasing in 1/f(alpha) noise spectra: origins, consequences, and remedies.
            Phys Rev E Stat Nonlin Soft Matter Phys 71, 066110 (2005).
    '''
    n = np.size(t)  # number of time points
    y = np.zeros(n)

    if f0 is None:
        f0 = 1/n  # fundamental frequency
    if m is None:
        m = n//2  # so the aliasing is limited

    k = np.arange(m) + 1  # wave numbers

    if seed is not None:
        np.random.seed(seed)

    theta = np.random.rand(int(m))*2*np.pi  # random phase

    f_vec = k*f0
    regime1= k*f0>=f_break
    regime2= k*f0<=f_break
    f_vec1 = f_vec[regime1]
    f_vec2 = f_vec[regime2]
    s = np.exp(alpha1/alpha2*np.log(f_vec1[0])) / f_vec2[-1]

    for j in range(n):
        coeff = np.ndarray((np.size(f_vec)))
        coeff[regime1] = f_vec1**(-alpha1/2)
        coeff[regime2] = (s*f_vec2)**(-alpha2/2)
        sin_func = np.sin(2*np.pi*k*f0*t[j] + theta)
        y[j] = np.sum(coeff*sin_func)

    return y
    
```

## Constant Frequency Signal

```python
freq=1/20
time=np.arange(2001)
signal=np.cos(2*np.pi*freq*time)

ts=pyleo.Series(time=time,value=signal)
fig=ts.plot(title='perfect signal',figsize=(8,2))
#plt.show()
fig1,fig2=compare_method(ts)
#plt.show()
```

## two mixed constant frequencies signal

```python
freq1=1/20
freq2=1/100
signal=np.cos(2*np.pi*freq1*time)+np.cos(2*np.pi*freq2*time)
ts=pyleo.Series(time=time,value=signal)
fig=ts.plot(title='2 close signals')
#plt.show()
fig1,fig2=compare_method(ts)
#plt.show()
```

## Colored Noise

```python
signal=colored_noise(1,time,seed=2333)
ts=pyleoclim.Series(time=time,value=signal)
ffig=ts.plot(title='colored noise')
plt.show()
fig1,fig2=compare_method(ts,signif=True)
plt.show()



#TODO: include signif testing
#https://github.com/LinkedEarth/LiPDbooks/blob/master/notebooks/pyleoclim_ui_tutorial.ipynb
```

## Colored Noise with 2 regimes

```python
signal=colored_noise_2regimes(1,2,1/100,time,seed=2333)
ts=pyleoclim.Series(time=time,value=signal)
fig=ts.plot(title='colored noise with 2 regimes')
plt.show()
fig1,fig2=compare_method(ts,signif=True)
plt.show()
```

## White Noise

```python
snr=0.25
freq=1/20
signal=np.cos(2*np.pi*freq*time)
sig_var=np.var(signal)
noise_var=sig_var/snr
white_noise=np.random.normal(0,np.sqrt(noise_var),size=np.size(signal))
signal+=white_noise
ts=pyleoclim.Series(time=time,value=signal)
fig1,fig2=ts.plot(title='white noise')
plt.show()
fig1,ax=compare_method(ts, signif=True)
plt.show()
```

## Signal with linear trend

```python
slope=0.01
intercept=-1
linear_trend=slope*time+intercept
freq=1/20

signal=np.cos(2*np.pi*freq*time)
signal+=linear_trend
ts=pyleoclim.Series(time=time,value=signal)
fig=ts.plot(title='constant signal with linear trend')
plt.show()
fig1,fig2=compare_method(ts)
plt.show()
```

## Signal with Nonlinear trend

```python
slope=1e-5
intercept=-1
nonlinear_trend=slope*time**2 + intercept
signal=np.cos(2*np.pi*freq*time)
signal+=nonlinear_trend
ts=pyleoclim.Series(time=time,value=signal)
fig=ts.plot(title='constant signal with nonlinear trend')
plt.show()
fig1,fig2=compare_method(ts)
plt.show()
```

## Randomly deleted data points

```python
del_percent=0.4
freq=1/20
signal=np.cos(2*np.pi*freq*time)
n_del=int(del_percent*np.size(time))
deleted_idx=np.random.choice(range(np.size(time)),n_del,replace=False)
time_uneven=np.delete(time,deleted_idx)
signal_uneven=np.delete(signal,deleted_idx)

ts=pyleoclim.Series(time=time_uneven,value=signal_uneven)
detrended_signal=ts.detrend(method='emd')
fig=ts.plot(title='constant signal with deleted points')
plt.show()
fig1,fig2=compare_method(ts,series2=detrended_signal)
plt.show()
```

## Deleted Gap

```python
nt=np.size(time)
freq=1/20
gaplength=200
signal=np.cos(2*np.pi*freq*time)
deleted_idx=np.arange(nt//2-gaplength//2,nt//2+gaplength//2)
time_gap=np.delete(time,deleted_idx)
signal_gap=np.delete(signal,deleted_idx)

ts=pyleoclim.Series(time=time_gap,value=signal_gap)
fig=ts.plot(title='constant signal with deleted gap')
plt.show()
fig1,ax=compare_method(ts)
plt.show()
```
