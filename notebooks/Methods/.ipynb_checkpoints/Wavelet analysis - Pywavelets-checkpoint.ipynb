# Wavelet Analysis - pywavelets

This notebook demonstrates the use of Pywavelets for cwt analysis to be incoportated into Pyleoclim.

```python
import numpy as np
import pywt
import pyleoclim as pyleo
```

```python
t = np.arange(2001)
f0 = 1/50
signal = np.cos(2*np.pi*f0*t)
ts = pyleo.Series(time=t, value=signal,time_unit='year')
```

```python
#plot
ts.plot()
```

```python
#WWZ analysis
scal = ts.wavelet(method='wwz')
```

```python
#plot
scal.plot()
```
